```python
import polars as pl
import playsummary
```

```python
top1000 = pl.read_csv('data/top1000.csv')
median_plays = []
```

```python
for rank, name, game_id in top1000.iter_rows():
    df = playsummary.data(game_id)
    plays = df['Plays per User'].median()
    median_plays.append(plays)
    print(f"{rank}\t{plays:.3f}")
```

```python
most_played = (
    top1000
        .with_columns(
            pl.Series("avg_plays", median_plays)
        )
        .sort("avg_plays", descending=True)
        .with_row_index("play_rank", offset=1)
        .rename({'rank': 'bgg_rank'})
)

most_played.write_csv('data/mostplayed.csv')
```

```python
with pl.Config(tbl_rows=25):
    display(most_played.limit(25))
    display(most_played.tail(5))
```
